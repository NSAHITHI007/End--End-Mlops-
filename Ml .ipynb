```python
new_dataset=dataset.drop(columns=['AveBedrms'])
```

```python
import seaborn as sns 
sns.pairplot(new_dataset)
```

```python
plt.scatter(new_dataset['AveRooms'], new_dataset['Price'])
plt.xlabel("Average Rooms")
plt.ylabel("Price")
```

```python
sns.regplot(x='AveRooms',y='Price',data=new_dataset)
```

```python
X = new_dataset.iloc[:, :-1]
Y = new_dataset.iloc[:, -1]
print(X.head())   
```

```python
print(Y.head())
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
X_test
```

Standard Scaler : same scale of the independent features

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train,Y_train)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
regression.get_params()
```

```python
reg_predict=regression.predict(X_test)
```

```python
reg_predict
```

```python
plt.scatter(Y_test,reg_predict)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

print("R² Score:", r2_score(Y_test, reg_predict))
print("MAE:", mean_absolute_error(Y_test, reg_predict))
```

```python
residuals=Y_test-reg_predict
residuals
```

```python
sns.displot(residuals,kind="kde")
```

```python
plt.scatter(reg_predict,residuals)
```

```python
import numpy as np

# Get the feature index of 'AveBedrms'
avebedrms_index = list(fetch_california_housing.feature_names).index('AveBedrms')

# Remove 'AveBedrms' from the first data sample
corrected_sample = np.delete(fetch_california_housing.data[0], avebedrms_index)

# Reshape for model input
corrected_sample = corrected_sample.reshape(1, -1)

# Make prediction
prediction = regression.predict(corrected_sample)

print("Predicted House Price:", prediction[0])
```
